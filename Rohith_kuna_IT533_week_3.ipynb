```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
kickstarter = pd.read_csv("https://raw.githubusercontent.com/shstreuber/Data-Mining/master/data/kickstarter-projects-201518.csv", encoding="ISO-8859-1") 
kickstarter.dropna(inplace=True)
```

```python
kickstarter.head()
```

```python
state_num=kickstarter.state

kickstarter['state_num'] = state_num

kickstarter.state_num
```

```python
kickstarter['state_num']= kickstarter['state_num'].astype('category')

kickstarter['state_num']= kickstarter['state_num'].cat.codes

kickstarter.dtypes[['state','state_num']]
```

```python
country_num=kickstarter.country

kickstarter['country_num'] = country_num

kickstarter.country_num
```

```python
kickstarter['country_num']= kickstarter['country_num'].astype('category')

kickstarter['country_num']= kickstarter['country_num'].cat.codes

kickstarter.dtypes[['country','country_num']]
```

```python
category_num=kickstarter.category

kickstarter['category_num'] = category_num

kickstarter.category_num
```

```python
kickstarter['category_num']= kickstarter['category_num'].astype('category')

kickstarter['category_num']= kickstarter['category_num'].cat.codes

kickstarter.dtypes[['category','category_num']]
```

1. Data Transformation.      
   Here in this data transformation we have alligned numbers to three different kind of variables so that the data can be arranged numerically and will be easy for the person to find the required data.

```python
kickstarter2 = kickstarter
kickstarter2.currency.unique()
```

```python
kickstarter2[kickstarter2['currency'] == 'USD']
```

```python
kickstarter2.currency.replace('USD','dollar', inplace=True)

kickstarter2[kickstarter2['currency'] == 'USD']
kickstarter2.currency.unique()
```

```python
kickstarter2 = kickstarter
kickstarter2.country.unique()
```

```python
kickstarter2[kickstarter2['country'] == 'US']
```

```python
kickstarter2.country.replace('US','united states', inplace=True)

kickstarter2[kickstarter2['country'] == 'US']
kickstarter2.country.unique()
```

2. Data substitution.      
  In this data substitution I've decided to change currency USD to dollars and country US to united states. I find it more accurate that if we change these we can search for them easily and we can cateorise them easily.

```python
bins = np.linspace(kickstarter.goal.min(),kickstarter.goal.max(),4)
bins
```

```python
labels = ['small', 'medium', 'high']
```

```python
kickstarter3=kickstarter
kickstarter3['bins'] = pd.cut(kickstarter3['goal'], bins=bins, labels=labels, include_lowest=True)
```

```python
kickstarter3.head()
```

```python
buckets = [ 1.00000000e+00, 2.00000008e+07, 4.00000006e+07, 6.00000004e+07,
       8.00000002e+07, 1.00000000e+08 ]
bucketlabels = ['poor', 'low-wage-earners', 'middle-class', 'rich', 'super-rich']

kickstarter3['buckets'] = pd.cut(kickstarter3['goal'] , bins=buckets, labels=bucketlabels, include_lowest=True)

kickstarter3.head()
```
